```python
book = pd.read_csv("books.csv", sep = ";",error_bad_lines = False,encoding = "latin-1")
```

```python
print(book.shape)
```

```python
book.drop(['Image-URL-S','Image-URL-M'],axis = 1,inplace = True)
```

```python
book["Year-Of-Publication"].unique()
```

```python
book.loc[book["Year-Of-Publication"] == 'DK Publishing Inc',:]
```

```python
book.loc[book["Year-Of-Publication"] == 'Gallimard',:]
```

```python
book = book[(book["Year-Of-Publication"] != 'DK Publishing Inc') & (book["Year-Of-Publication"] != 'Gallimard')]
```

```python
book["Year-Of-Publication"] = book["Year-Of-Publication"].astype('int32')
book.dtypes
```

```python
book["Publisher"].isnull().sum()
```

```python
book = book.dropna(subset = ["Publisher"])
```

```python
book["Publisher"].isnull().sum()
```

```python
book["Book-Title"].isnull().sum()
```

```python
book["Book-Author"].isnull().sum()
```

```python
book = book.dropna(subset = ["Book-Author"])
```

```python
book.head()
```

```python
book['Year-Of-Publication'] = book['Year-Of-Publication'].apply(str)
```

```python
book.dtypes
```

```python
book["Book-Author"] = book["Book-Author"].apply(lambda x:x.split())
```

```python
book["Book-Title"] = book["Book-Title"].apply(lambda x:x.split())
```

```python
book["Publisher"] = book["Publisher"].apply(lambda x:x.split())
```

```python
book['Year-Of-Publication'] = book['Year-Of-Publication'].apply(lambda x:x.split())
```

```python
book['tags'] = book["Book-Author"] + book["Book-Title"] + book["Publisher"] + book['Year-Of-Publication']
```

```python
book['tags'][0]
```

```python
book.head(1)
```

```python
new_book = book[['ISBN','Book-Title','tags']]
```

```python
new_book.head()
```

```python
new_book['tags'] = new_book['tags'].apply(lambda x: " ".join(x))
```

```python
new_book['tags'][0]
```

```python
new_book['tags'] = new_book['tags'].apply(lambda x:x.lower())
```

```python
new_book.head()
```

```python
new_book['Book-Title'] = new_book['Book-Title'].apply(lambda x: " ".join(x))
```

```python
new_book.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 2000,stop_words='english')
```

```python
vector_book = cv.fit_transform(new_book['tags']).toarray()
```

```python
# using steming to get {dance,dancing,danced} to {danc,dance,dance}
#nltk library
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
     
    return " ".join(y)
        
```

```python
new_book['tags'].apply(stem)
```
